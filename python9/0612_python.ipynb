```python
import pandas as pd
import matplotlib.pyplot as plt


# 데이터 파일 경로
data_path = '일별평균대기오염도_2022.csv'

# CSV 파일 읽기
df = pd.read_csv(data_path, encoding="cp949")

# 필요한 필드 추출
df = df[['측정일시', '측정소명', '미세먼지농도(㎍/㎥)', '초미세먼지농도(㎍/㎥)']]

# 결측치 처리
df = df.dropna()

# 측정일시 컬럼의 데이터 타입을 datetime으로 변경
df['측정일시'] = pd.to_datetime(df['측정일시'], format='%Y%m%d')

# 연도별 미세먼지와 초미세먼지 농도 평균 계산
df_monthly = df.resample('M', on='측정일시').mean(numeric_only=True)

# 그래프 그리기
plt.plot(df_monthly.index.month, df_monthly['미세먼지농도(㎍/㎥)'], label='PM10')
plt.plot(df_monthly.index.month, df_monthly['초미세먼지농도(㎍/㎥)'], label='PM2.5')
plt.legend()
plt.xlabel('Month')
plt.ylabel('Concentration')
plt.rc('font', family='Malgun Gothic')
plt.title('일별평균대기오염도')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

years = np.arange(2010, 2021)  # 연도 데이터
gdp_a = np.array([100, 120, 150, 160, 180, 200, 220, 240, 260, 280, 300])  # A 나라의 연도별 GDP
gdp_b = np.array([80, 90, 100, 110, 120, 130, 140, 150, 160, 170, 180])  # B 나라의 연도별 GDP
gdp_c = np.array([200, 220, 240, 250, 260, 270, 280, 290, 300, 310, 320])  # C 나라의 연도별 GDP
index = np.arange(len(years))
bar_width = 0.2  # Narrower bar width

plt.bar(index, gdp_a, width=bar_width, label='A나라 GDP')
plt.bar(index+bar_width, gdp_b, width=bar_width, label='B나라 GDP')
plt.bar(index+2*bar_width, gdp_c, width=bar_width, label='C나라 GDP')

plt.rc('font', family='Malgun Gothic')
plt.xlabel('년도')
plt.ylabel('GDP')
plt.title('나라별 GDP')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

years = np.arange(2010, 2021)
gdp = np.array([100, 120, 150, 160, 180, 200, 220, 240, 260, 280, 300])
sales = np.array([50, 70, 30, 45, 60, 80, 70, 90, 110, 100, 120])
prices = np.array([10, 12, 15, 16, 18, 20, 22, 24, 26, 28, 30])

plt.figure(figsize=(8, 6))

# Plot GDP as a line graph
plt.plot(years, gdp, marker='o', label='GDP')

plt.rc('font', family='Malgun Gothic')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('GDP Trend')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))

# Plot sales as a bar graph
plt.bar(years, sales, label='Sales')

plt.rc('font', family='Malgun Gothic')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Performance')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))

plt.plot(years, sales, marker='o', label='GDP')

plt.rc('font', family='Malgun Gothic')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Performance')
plt.legend()
plt.grid(True)
plt.show()

# Plot prices as scatter plot points
plt.scatter(years, prices, color='green', label='Prices')

plt.rc('font', family='Malgun Gothic')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Price Fluctuations')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# 예시 데이터프레임 생성
data = {'학과': ['전산학과', '전자공학과', '전자공학과', '경영학과', '경영학과'],
        '성적': [80, 90, 75, 85, 95],
        '인원': [50, 40, 35, 60, 55]}
df = pd.DataFrame(data)

# '학과' 열을 기준으로 그룹화한 후 '성적' 열에 대해 평균과 합계 계산
grouped = df.groupby('학과')  # '학과' 열을 기준으로 그룹화

summary_stats = grouped['성적'].agg(['mean', 'sum'])  # 그룹별 '성적' 열의 평균과 합계 계산
print(summary_stats)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 데이터 파일 경로
data_path = '일별평균대기오염도_2022.csv'

# CSV 파일 읽기
df = pd.read_csv(data_path, encoding="cp949")

# 필요한 필드 추출
df = df[['측정일시', '측정소명', '미세먼지농도(㎍/㎥)', '초미세먼지농도(㎍/㎥)']]

# 결측치 처리
df = df.dropna()

# 측정일시 컬럼의 데이터 타입을 datetime으로 변경
df['측정일시'] = pd.to_datetime(df['측정일시'], format='%Y%m%d')

# 일별 합계 계산
df_daily = df.resample('D', on='측정일시').sum(numeric_only=True)

# 일평균 대기오염도 계산
df_daily['미세먼지농도(㎍/㎥)'] /= 24
df_daily['초미세먼지농도(㎍/㎥)'] /= 24

# 그래프 그리기
plt.plot(df_daily.index, df_daily['미세먼지농도(㎍/㎥)'], label='PM10')
plt.plot(df_daily.index, df_daily['초미세먼지농도(㎍/㎥)'], label='PM2.5')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Concentration')
plt.title('2022 Air Pollution Trend')
plt.show()
```
