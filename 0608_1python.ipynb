```python
class Deposit:
    def __init__(self, initial, interest, n):
        self.initial = initial
        self.interest = interest
        self.n = n

    def profit(self):
        total = self.initial * (1 + self.interest) ** self.n
        return int(total)

initial_money = 1000000
interest_percent = 0.035
years = 7

deposit = Deposit(initial_money, interest_percent, years)
final_amount = deposit.profit()

print(final_amount, "원")

```

```python
# 예제의 Teacher 클래스에서 People 클래스의 __init__()를 호출하지 않고 부모 클래스의 age, name 인스턴스 변수를 이용할 수 있는가? 이용할 수 없다면 그 이유는? 
# 이용할 수 있게 하려면 프로그램을 어떻게 수정해야 하는가?

## 예제의 Teacher 클래스에서 People 클래스의 __init__()를 호출하지 않고 부모 클래스의 age와 name 인스턴스 변수를 사용할 수 없습니다. 
## 이는 Teacher 클래스에서 __init__() 메소드를 재정의(override)하여 People 클래스의 초기화 함수를 호출하지 않고 있기 때문

class People:
    def __init__(self, age=0, name=None):
        self.__age = age
        self.__name = name

    def introMe(self):
        print("Name:", self.__name, "Age:", str(self.__age))


class Teacher(People):
    def __init__(self, age=0, name=None, school=None):
        super().__init__(age, name)  # 부모 클래스의 __init__() 호출
        self.school = school

    def showSchool(self):
        print("My School is", self.school)
```

```python
# 다음 Person 클래스를 상속 받는 Employee 클래스를 정의하라. 
# Employee 클래스에 employeeID 인스턴스 변수를 추가하고 getID() 메소드를 정의하라. 
# getID() 메소드는 employeeID를 반환하는 메소드이다. 
# Employee 클래스를 이용하여 Employee("동양", 65, 2019)로 생성된 객체의 이름, 나이, ID를 출력하라.

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def get_name(self):
        return self.name

    def get_age(self):
        return self.age


class Employee(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

    def get_id(self):
        return self.employee_id


employee = Employee("동양", 65, 2019)
name = employee.get_name()
age = employee.get_age()
id = employee.get_id()

print("이름:", name)
print("나이:", age)
print("ID:", id)
```

```python
class Student:
    def __init__(self, name, student_id, year, major, avg_score):
        self.name = name
        self.student_id = student_id
        self.year = year
        self.major = major
        self.avg_score = avg_score

    def get_info(self):
        return f"이름: {self.name}\n학번: {self.student_id}\n학년: {self.year}\n전공: {self.major}\n평균 성적: {self.avg_score}"


class StudentManager:
    def __init__(self):
        self.student_list = []

    def add_student(self, student):
        self.student_list.append(student)

    def remove_student(self, student_id):
        for student in self.student_list:
            if student.student_id == student_id:
                self.student_list.remove(student)
                print(f"학번 {student_id}인 학생을 제거하였습니다.")
                return
        print(f"학번 {student_id}인 학생을 찾을 수 없습니다.")

    def find_student(self, student_id):
        for student in self.student_list:
            if student.student_id == student_id:
                return student
        return None

    def show_all_students(self):
        if len(self.student_list) == 0:
            print("등록된 학생이 없습니다.")
            return
        print("전체 학생 정보:")
        for student in self.student_list:
            print(student.get_info())


# 학생 객체 생성
student1 = Student("양성부", "20210001", 2, "컴퓨터공학", 85.7)
student2 = Student("손주원", "20210002", 3, "사회복지학", 92.3)
student3 = Student("김도경", "20210003", 1, "전자공학", 78.9)

# 학생들을 관리하는 객체 생성
manager = StudentManager()

# 학생 추가
manager.add_student(student1)
manager.add_student(student2)
manager.add_student(student3)

# 모든 학생 정보 출력
manager.show_all_students()

# 학번을 이용해 학생 찾기
found_student = manager.find_student("20210002")
if found_student:
    print("\n찾은 학생 정보:")
    print(found_student.get_info())

# 학번을 이용해 학생 제거
manager.remove_student("20210003")

# 모든 학생 정보 출력
manager.show_all_students()
```
